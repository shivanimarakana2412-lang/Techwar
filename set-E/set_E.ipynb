```python
df["Region"].fillna(df["Region"].mode())
df["City"].fillna(df["City"].mode()[0],inplace=True)
df["Course_Name"] = df["Course_Name"].fillna(df["Course_Name"].mode()[0])
df["Course_Status"] = df["Course_Status"].fillna(df["Course_Status"].mode()[0])
df["Placement_Status"] = df["Placement_Status"].fillna("Non")
df["Attendance_Percent"] = df["Attendance_Percent"].fillna(df["Attendance_Percent"].median())
```

# TASK 4

```python
# Bar chart
course_count = df['Course_Name'].value_counts()

plt.bar(course_count.index, course_count.values)
plt.title('Course vs Enrollment Count')
plt.xlabel('Course Name')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.show()
```

```python
# Line chart
df['Enrollment_Date'] = pd.to_datetime(df['Enrollment_Date'])

enroll_trend = df.groupby(df['Enrollment_Date'].dt.year).size()

plt.plot(enroll_trend.index, enroll_trend.values, marker='o')
plt.title('Enrollment Trend Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Enrollments')
plt.show()
```

```python
# Histo gram
plt.hist(df['Attendance_Percent'], bins=10)
plt.title('Attendance Percentage Distribution')
plt.xlabel('Attendance Percentage')
plt.ylabel('Number of Students')
plt.show()
```

```python
# scatter plot
plt.scatter(df['Attendance_Percent'], df['Fees_Paid'], color = 'pink'  , edgecolors = 'black' , s = 100)
plt.title('Attendance vs Fees Paid')
plt.xlabel('Attendance Percentage')
plt.ylabel('Fees Paid')
plt.show()
```

```python
# Pie chart
learning_mode_count = df['Learning_Mode'].value_counts()

plt.pie(learning_mode_count.values, labels=learning_mode_count.index, autopct='%1.1f%%')
plt.title('Learning Mode Distribution')
plt.show()
```

```python
# Sub plot
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.bar(course_count.index, course_count.values)
plt.title('Course Enrollment')
plt.xticks(rotation=45)

plt.subplot(1,2,2)
plt.hist(df['Attendance_Percent'], bins=10)
plt.title('Attendance Distribution')

plt.tight_layout()
plt.show()
```
