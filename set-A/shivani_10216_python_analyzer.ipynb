```python
print("Category wise performance:-")
category_performance = df.groupby('Product_Category')['Quantity'].sum().sort_values(ascending=False)
print(category_performance)
```

```python
pd.DataFrame(df)
```

```python
summary = df.groupby("Product_Category")["Revenue"].sum()

plt.figure(figsize=(8,4))
plt.bar(summary.index, summary.values)
plt.title("Revenue by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(x ="Order_Date", y = "Revenue", marker='o')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.hist(df["Revenue"], bins=5)
plt.title("Revenue Distribution")
plt.xlabel("Revenue Range")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(df["Revenue"], labels=df["Product_Category"], autopct='%1.1f%%')
plt.title("Revenue Share by Category")
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.scatter(df["Quantity"], df["Revenue"])
plt.title("Quantity vs Revenue")
plt.xlabel("Quantity Sold")
plt.ylabel("Revenue")
plt.show()
```

# task 5

```python


df = pd.DataFrame(df)

# --- Key Findings ---
top_category = df.groupby("Category")["Revenue"].sum().idxmax()
top_region = df.groupby("Region")["Revenue"].sum().idxmax()
top_customer = df.groupby("Customer_Type")["Revenue"].sum().idxmax()
best_month = df.groupby("Month")["Revenue"].sum().idxmax()
best_discount = df.groupby("Discount")["Revenue"].mean().idxmax()

print(" FINAL INSIGHTS")
print("----------------------------")
print(f"Highest revenue category      : {top_category}")
print(f"Top performing region         : {top_region}")
print(f"Best customer type            : {top_customer}")
print(f"Highest sales month           : {best_month}")
print(f"Most effective discount (%)   : {best_discount}")
```
